###### status[0] is the 1st person, X[1] is the 2nd person,
###### X[i,0],X[i,1],X[i,2] is the storage of food, cloth, and money of the i-th person

```python
import numpy as np
status = np.ones((2, 3))*10;
print(status)
```

```python
consumption = [1,0.1,0];
production = np.zeros((2, 3));
production[0,0] = 2.2;
production[1,1] = 0.3;
conductivity = [1,0.3,0]; # how the desire transformed into movement
Status = [status];
ifp = 1; # controller to show the detail of steps
for i in range(100):
    desire = Status[-1][0]-Status[-1][1]; # desire to trade goods, difference of storage
    if ifp == 0:
        print("desire", desire, end=' ');
    trade=desire * conductivity;
    if ifp == 0:
        print("trade", trade);
    exchange = np.array([trade*-1,trade*1]);
    newstatus = Status[-1] - consumption + production + exchange;
    Status.append(np.round(newstatus, 2));
final=np.array(Status);
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
x=range(0,len(final))
# plt.gca().set_aspect('equal')
for i in range(len(final[0])):
    for j in range(len(final[0,1])):
        plt.plot(x, final[x,i,j],label = np.array2string(np.array([i,j])))
plt.xlabel('Day')
plt.ylabel('status')
plt.legend(loc='best')
plt.grid()
plt.show()
```
